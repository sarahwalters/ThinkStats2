EXERCISE 5.1

```python
import scipy.stats
```

For example <tt>scipy.stats.norm</tt> represents a normal distribution.

```python
mu = 178
sigma = 7.7
dist = scipy.stats.norm(loc=mu, scale=sigma)
type(dist)
```

A "frozen random variable" can compute its mean and standard deviation.

```python
dist.mean(), dist.std()
```

It can also evaluate its CDF.  How many people are more than one standard deviation below the mean?  About 16%

```python
dist.cdf(mu-sigma)
```

How many people are between 5'10" and 6'1"? About 34%

```python
under_5ft10 = dist.cdf(177.80)
under_6ft1 = dist.cdf(185.42)
under_6ft1-under_5ft10
```

<tt>scipy.stats.pareto</tt> represents a pareto distribution.  In Pareto world, the distribution of human heights has parameters alpha=1.7 and xmin=1 meter.  So the shortest person is 100 cm and the median is 150.

```python
alpha = 1.7
xmin = 1
dist = scipy.stats.pareto(b=alpha, scale=xmin)
dist.median()
```

What is the mean height in Pareto world?

```python
dist.mean()
```

What fraction of people are shorter than the mean? 77.9%

```python
dist.cdf(dist.mean())
```

Out of 7 billion people, how many do we expect to be taller than 1 km?  You could use <tt>dist.cdf</tt> or <tt>dist.sf</tt>.

```python
dist.sf(1000)*7e9
```

EXERCISE 5.3

```python
import random
import thinkstats2
import thinkplot
```

```python
weibull_rand = [random.weibullvariate(0.5,1) for i in range(1000)]
weibull_cdf = thinkstats2.Cdf(weibull_rand, label='weibull')
thinkplot.Cdf(weibull_cdf, complement=True)
thinkplot.Show(yscale='log')
```

Slope is -k/lam

Make a random selection from <tt>cdf</tt>.

```python
weibull_cdf.Random()
```

Draw a random sample from <tt>cdf</tt>.

```python
weibull_rand = weibull_cdf.Sample(1000)
weibull_rand
```

Draw a random sample from <tt>cdf</tt>, then compute the percentile rank for each value, and plot the distribution of the percentile ranks. * Turns out as a linear distribution - as expected

```python
percentileRanks = [weibull_cdf.PercentileRank(x) for x in weibull_rand]
percentileDistr = thinkstats2.Cdf(percentileRanks)
thinkplot.Cdf(percentileDistr)
thinkplot.Show()
```
