```python
import nsfg
import thinkstats2
import thinkplot
import numpy as np
```

```python
df = nsfg.ReadFemPreg()
df = df.dropna(subset=['ager', 'totalwgt_lb'])

ages = df.ager
wgts = df.totalwgt_lb

thinkplot.Scatter(ages, wgts, alpha=0.2)
thinkplot.Show()
```

```python
thinkstats2.Corr(ages, wgts)
```

```python
thinkstats2.SpearmanCorr(ages, wgts)
```

These variables aren't strongly correlated.
