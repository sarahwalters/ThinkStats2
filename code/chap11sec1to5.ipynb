```python
t=[]
for name in join.columns:
    try:
        if join[name].var() < 1e-7:
            continue
        
        formula = str('totalwgt_lb ~ agepreg + ' + name)
        model = smf.ols(formula, data=join)
        if model.nobs < len(join)/2: # number of observations for model
            continue

        results = model.fit()
    except (ValueError, TypeError):
        continue
    
    t.append((results.rsquared, name))
```

Simple logistic regression example:

```python
import numpy as np
y = np.array([0, 1, 0, 1])
x1 = np.array([0, 0, 0, 1])
x2 = np.array([0, 1, 1, 1])

beta = [-1.5, 2.8, 1.1]
log_o = beta[0] + beta[1]*x1 + beta[2]*x2
log_o
```

```python
o = np.exp(log_o)
o
```

```python
p = o / (o+1)
p
```

```python
# likelihood is p when y==1 and (1-p) when y==0
likelihoods = (y*p) + ((1-y)*(1-p))
likelihoods
```

```python
likelihood = np.prod(likelihoods)
likelihood # goal of logistic regression is to minimize this - cost fn
```

```python
df = live[live.prglngth>30]
df['boy'] = (df.babysex==1).astype(int)
model = smf.logit('boy ~ agepreg', data=df) # Logit object
results = model.fit() # BinaryResults object
results.summary()
```

```python
import pandas
endog = pandas.DataFrame(model.endog, columns=[model.endog_names])
exog = pandas.DataFrame(model.exog, columns=model.exog_names)
# endog = dependent; exog = independent
# sometimes useful to have these as pandas dfs
```

```python
# Create model (better than just agepreg)
formula = 'boy ~ agepreg + hpagelb + birthord + C(race)'
model = smf.logit(formula, data=df)
results = model.fit()
```

```python
# this again, for new model
endog = pandas.DataFrame(model.endog, columns=[model.endog_names])
exog = pandas.DataFrame(model.exog, columns=model.exog_names)
```

```python
# Accuracy of model
predict = (results.predict() >= 0.5)
actual = endog['boy']
true_pos = predict*actual
true_neg = (1-predict) * (1-actual)
acc = (sum(true_pos) + sum(true_neg))/len(actual)
acc
```
